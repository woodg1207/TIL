### 연습문제 서류전형 합격여부

```python
df = pd.read_csv('data/toeic.csv')
df
```

```python
(df['LC']>250 ) & (df['RC']>250) &(df['LC']+df['RC']>600)
```

```python
df['합격 여부'] = False
```

```python
df.loc[(df['LC']>250 ) & (df['RC']>250) &(df['LC']+df['RC']>600),'합격 여부'] = True
```

```python
df
```

```python
## 해설
df = pd.read_csv('data/toeic.csv')

pass_total = df['LC'] + df['RC'] > 600
pass_both = (df['LC'] >= 250) & (df['RC'] >= 250)
df['합격 여부'] = pass_total & pass_both
df
```
