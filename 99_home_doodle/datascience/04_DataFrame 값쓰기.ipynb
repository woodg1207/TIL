```python
## 값 추가 삭제
ip_df = pd.read_csv('./data/iphone.csv', index_col=0)
```

```python
## 없는 Row 나 column을 넣으면 새로운 정보가 생성된다.
ip_df.loc['samsung'] = ['0305','5.5','6GB','And 10', 'Yes']
ip_df['제조사'] = '?'
ip_df
```

```python
## 값 삭제
ip_df.drop('samsung', axis='index', inplace=False ) 
# inplace=False 기존의 데이터 프레임을 건드리지 않는다. 계속사용하려면 변수에 저장, 반대는 True
```

```python
ip_df
```

```python
ip_df.drop('제조사', axis='columns', inplace=True)
```

```python
ip_df
```

```python
###  여러개 지울때
ip_df.drop(['samsung', 'iPhone 7'], axis='index', inplace=False ) 
```
